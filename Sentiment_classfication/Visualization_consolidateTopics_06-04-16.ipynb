```python
# Load the main topic map that was created for the visualization.
prefix = "../data/visualization/"
#file = "TopicsAndSubtopics_clean2.csv"
#file = "TopicsAndSubtopics_clean_05-31-16.csv"
file = "TopicsAndSubtopics_clean_06-04-16.csv"
df = pd.read_csv(prefix + file)
print df.shape
df.head(2)
```

```python
# Get list of all raw topics in the main topic map.
rawTopics = set(df['RawTopic'].tolist())
print type(rawTopics)
print len(rawTopics)
```

```python
# Locate the labeled, timestamped transcipt data that was grouped by candidate and topic.
topicData_prefix = "../data/topic/"
%ls ../data/topic/
```

```python
# List the files to load.
debateFiles = ['D1_G1_timeDurations.csv', 'D5_G3_timeDurations.csv',
                    'D6_G4_timeDurations.csv', 'D7_G6_timeDurations.csv',
                    'D8_G6_timeDurations.csv', 
                    'R10_G5_timeDurations.csv','R11_G6_timeDurations.csv', 
                    'R7_G2_timeDurations.csv','R8_G3_timeDurations.csv', 
                    'R9_G4_timeDurations.csv']
print debateFiles
```

```python
# For each file, load it and see which of its topics are not yet in the main topic mapping file.
dataframes = []
for debateFile in debateFiles:
    print debateFile
    df_debate = pd.read_csv( topicData_prefix + debateFile )
    dataframes.append( df_debate )
    print df_debate.shape
    df_debate.head(2)

    # Get list of all topics for this debate.
    debateTopics = set(df_debate['mappedTopicPreeti'].tolist())
    print len(debateTopics), "topics"

    # Get set difference. If empty, that means all topics for this debate are already in the main topic map.
    print debateTopics-rawTopics
    print
print len(dataframes)
print 'done'
```

```python
# Load the debate calendar.
df_calendar = pd.read_csv("../data/visualization/DebateCalendar.csv")
df_calendar.drop(['Unnamed: 5'],inplace=True,axis=1)
df_calendar
```

```python
# Consolidate the files into one.
df_all = pd.concat( dataframes )
print df_all.shape
df_all.head(2)
```

```python
# Assign date to each debate.
df_calendar = df_calendar.rename(columns = {'Event':'event'})
df_all = pd.merge(df_all, df_calendar, on='event', how='left')
df_all.head(2)
```
