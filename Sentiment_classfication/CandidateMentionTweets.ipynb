```python
candidates = ['Trump','Clinton','Sanders','Cruz','Rubio']
```

```python
query = {}

query['Sanders'] = """Match (t:Tweet_{0}) \
where exists(t.mentions_Sanders) \
return t.id_str, t.text"""

query['Clinton'] = """Match (t:Tweet_{0}) \
where exists(t.mentions_Clinton) \
return t.id_str, t.text"""

query['Trump'] = """Match (t:Tweet_{0}) \
where exists(t.mentions_Trump) \
return t.id_str, t.text"""

query['Cruz'] = """Match (t:Tweet_{0}) \
where exists(t.mentions_Cruz) \
return t.id_str, t.text"""

query['Rubio'] = """Match (t:Tweet_{0}) \
where exists(t.mentions_Rubio) \
return t.id_str, t.text"""

# and not exists(t.mentions_Clinton) \
# and not exists(t.mentions_Rubio) \
# and not exists(t.mentions_Trump) \
# and not exists(t.mentions_Cruz) \
# and not exists(t.mentions_Kasich) \
# and not exists(t.mentions_Bush) \
# and not exists(t.mentions_Christie) \
# and not exists(t.mentions_Huckabee) \
# and not exists(t.mentions_Omally) \
# and not exists(t.mentions_Randpaul) \
# and not exists(t.mentions_Fiorina) \
# and not exists(t.mentions_Carson) \
```

```python
from neo4j.v1 import GraphDatabase, basic_auth
import time
import csv

print 'opening driver...'
driver = GraphDatabase.driver("bolt://127.0.0.1")
print 'after opening driver...'


for partition in partitions:
#for partition in ['2015_10_11']:

    for candidate in candidates:

        filename = 'candidatementions_all/' + candidate + '/' + partition + '.csv'

        print filename

        with open(filename, 'wb') as csvfile:
            
            tweetwriter = csv.writer(csvfile)

            start = time.time()

            queryformatted = query[candidate].format(partition)
            print queryformatted

            session = driver.session()
            result = session.run(queryformatted)

            for record in result:
                tweetwriter.writerow([record[0],record[1].encode("utf-8")])

            session.close()

            end = time.time()
            print end - start

        
```
