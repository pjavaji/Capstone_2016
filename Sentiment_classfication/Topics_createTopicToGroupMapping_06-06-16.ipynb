```python
# Load file containing the combined data from the debate transcripts.
prefix = "../data/visualization/"
filename = "All_timeDurations_3.csv"
df = pd.read_csv(prefix + filename)
print df.shape
df.head(2)
```

```python
# Group by event and mappedTopicPreeti.  Sum the timeDurationSec. Divide by 60 to get time duration in minutes.
df2 = df.groupby(['event', 'mappedTopicPreeti'])['timeDurationSec'].sum().reset_index()
df2['timeDurationMinutes'] = df2['timeDurationSec']/60
df2.drop('timeDurationSec', axis=1, inplace=True)
print df2.shape
#df2.sort(['timeDurationMinutes'], ascending=False)
df2.sort(['event'], ascending=True)
```

```python
# Write results to file.
outfilename = "DebateToTopicMapping_06-07-16.csv"
df2.to_csv(prefix+outfilename)
print "Wrote to file {}".format(prefix+outfilename)
```
