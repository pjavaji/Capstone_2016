```python
def headerpath(path, nodename):
    return "./" + path + "/" + nodename + "-header.csv"

def filepath(path, nodename):
    return "./" + path + "/" + nodename + "_dedup.csv"

    
```

```python
# Load csv data into a neo4j database

%cd ~/documents/DSE/dhofman/capstone/archive/output/
!ls -la

    
nodelist = ""
rellist = ""

for path in folderNames:
    
    print path
    
    for nodename in ["Hashtag", "Place", "Source", "Link", "Tweet", "User"]:
        nodelist = nodelist + "--nodes '" + ",".join([headerpath(path,nodename),filepath(path,nodename)]) + "' \\\n"

    for relname in ["CONTAINS", "LOCATED_IN", "RETWEETS","USING","REPLY_TO","TAGS","POSTS","MENTIONS"]:
        rellist = rellist + "--relationships '" + ",".join([headerpath(path,relname),filepath(path,relname)]) + "' \\\n"
    
lines = nodelist+rellist
lines = lines[:-2]
print lines
    
    
!~/documents/neo4j-community-2.3.3/bin/neo4j-import --multiline-fields=true \
--skip-bad-relationships --skip-duplicate-nodes \
--into ~/documents/DSE/dhofman/capstone/tweetloader/alldates \
$lines
#     --nodes "./2016_3_9/Hashtag-header.csv,./2016_3_9/Hashtag_dedup.csv" \
#     --nodes "./2016_3_9/Place-header.csv,./2016_3_9/Place_dedup.csv" \
#     --nodes "./2016_3_9/Source-header.csv,./2016_3_9/Source_dedup.csv" \
#     --nodes "./2016_3_9/Link-header.csv,./2016_3_9/Link_dedup.csv" \
#     --nodes "./2016_3_9/Tweet-header.csv,./2016_3_9/Tweet_dedup.csv" \
#     --nodes "./2016_3_9/User-header.csv,./2016_3_9/User_dedup.csv" \
#     --relationships "./2016_3_9/CONTAINS-header.csv,./2016_3_9/CONTAINS_dedup.csv" \
#     --relationships "./2016_3_9/LOCATED_IN-header.csv,./2016_3_9/LOCATED_IN_dedup.csv" \
#     --relationships "./2016_3_9/RETWEETS-header.csv,./2016_3_9/RETWEETS_dedup.csv" \
#     --relationships "./2016_3_9/USING-header.csv,./2016_3_9/USING_dedup.csv" \
#     --relationships "./2016_3_9/REPLY_TO-header.csv,./2016_3_9/REPLY_TO_dedup.csv" \
#     --relationships "./2016_3_9/TAGS-header.csv,./2016_3_9/TAGS_dedup.csv" \
#     --relationships "./2016_3_9/POSTS-header.csv,./2016_3_9/POSTS_dedup.csv" \
#     --relationships "./2016_3_9/MENTIONS-header.csv,./2016_3_9/MENTIONS_dedup.csv" 
```
